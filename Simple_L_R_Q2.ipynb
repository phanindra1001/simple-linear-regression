```python
model = smf.ols('Salary~YearsExperience',data=data).fit()
```

```python
model.params
```

```python
model.summary()
```

Here R^2 value is 0.957 so the model is good model.

```python
newdata = pd.Series([7])
pred_data = pd.DataFrame(newdata,columns=['YearsExperience'])
model.predict(pred_data)
```
